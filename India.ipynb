#**Exploratory Data Analysis India 2024 Statistics - EDA & Visualisation**

#Importing relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

#Loading the dataset

```python
data = pd.read_csv('/content/drive/MyDrive/state wise pop.csv')
data.replace('-','0')
```

```python
data.dtypes
```

```python
def check_unique_values(data):
    categorical_columns = ['States/Uts', 'Majority']
    for col in categorical_columns:
        print(f"Unique values in {col}:\n", data[col].unique())
check_unique_values(data)
```

#Data Cleaning

```python
columns_to_convert = ['population(1901)', 'population(1951)', 'population(2011)', 'population(2023)', 'population(2024)']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

# Population Analysis

```python
data.plot(x='States/Uts', y=['population(1901)', 'population(1951)', 'population(2011)', 'population(2023)', 'population(2024)'], kind='bar', figsize=(15,8))
plt.title('Population Growth Trends')
plt.xlabel('States/UTs')
plt.ylabel('Population')
plt.legend(['1901', '1951', '2011', '2023', '2024'])
plt.xticks(rotation=90)
plt.show()
```

```python
data['Density (1901)'] = data['population(1901)'] / data['Area (sq. km)']
data['Density (1951)'] = data['population(1951)'] / data['Area (sq. km)']
data['Density (2011)'] = data['population(2011)'] / data['Area (sq. km)']
data['Density (2023)'] = data['population(2023)'] / data['Area (sq. km)']
data['Density (2024)'] = data['population(2024)'] / data['Area (sq. km)']
ax=data.plot(x='States/Uts', y=['Density (1901)', 'Density (1951)','Density (2011)','Density (2023)','Density (2024)'], kind='line', figsize=(15, 8))
ax.set_xticks(np.arange(len(data)))
ax.set_xticklabels(data['States/Uts'], rotation=90)
ax.set_title('Population Density (2024)')
ax.set_xlabel('States/UTs')
ax.set_ylabel('Density (per sq. km)')
plt.show()
```

# Literacy Rate Analysis

```python
ax =data.plot(x='States/Uts', y=['Male(literacy rate)', 'Female (literacy rate)Average (literacy rate)', 'average (literacy rate)'], kind='line', figsize=(15,8), marker='+')
ax.set_xticks(np.arange(len(data)))
ax.set_xticklabels(data['States/Uts'], rotation=90)
ax.set_title('Literacy Rate Trends')
ax.set_xlabel('States/UTs')
ax.set_ylabel('Literacy Rate (%)')
ax.legend(['Male', 'Female', 'Average'])
plt.show()
```

# Economical Analysis

```python
fig, ax1 = plt.subplots(figsize=(15, 8))

color = 'tab:blue'
ax1.set_xlabel('States/UTs')
ax1.set_ylabel('GDP (in trillion INR)', color=color)
ax1.bar(data['States/Uts'], data['GDP'], color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(data['States/Uts'], rotation=90)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Per Capita GDP (INR)', color=color)
ax2.plot(data['States/Uts'], data['per capita in'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('GDP and Per Capita GDP Charts')
plt.show()
```

```python
ax=data.plot(x='States/Uts', y=['population(1901)', 'population(1951)', 'population(2011)', 'population(2023)', 'population(2024)'], kind='line', figsize=(15, 8))
ax.set_xticks(np.arange(len(data)))
ax.set_xticklabels(data['States/Uts'], rotation=90)
ax.set_title('Population Growth Charts')
ax.set_xlabel('States/UTs')
ax.set_ylabel('Population')
plt.legend(['1901', '1951', '2011', '2023', '2024'])
plt.show()
```

```python
regions = {
    'North': ['Haryana', 'Himachal Pradesh', 'Punjab', 'Uttarakhand', 'Uttar Pradesh'],
    'South': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana'],
    'East': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal'],
    'West': ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan'],
    'Northeast': ['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Tripura']
}

data['Region'] = data['States/Uts'].map(lambda x: next((k for k, v in regions.items() if x in v), 'Other'))

regional_data = data.groupby('Region').agg({
    'population(2024)': 'sum',
    'average (literacy rate)': 'mean',
    'GDP': 'sum',
    'sex ratio (number of female per male)': 'mean'
}).reset_index()

regional_data.plot(x='Region', kind='bar', subplots=True, layout=(2, 2), figsize=(15, 10), legend=False)
plt.suptitle('Regional Analysis on Various Metrics')
plt.tight_layout()
plt.show()
```

# Demographical Analysis

```python
data.plot(x='States/Uts', y=['Hindu', 'Muslim', 'Christian', 'Sikhs', 'Buddhist', 'Others'], kind='bar', stacked=True, figsize=(13, 6))
plt.title('Religious Demographics')
plt.xlabel('States/UTs')
plt.ylabel('Proportion of Population')
plt.xticks(rotation=90)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Plotting the pie chart
religion_columns = ['Hindu', 'Muslim', 'Christian', 'Sikhs', 'Buddhist', 'Others']
religion_sums = data[religion_columns].sum()
plt.figure(figsize=(10, 7))
plt.pie(religion_sums, labels=religion_sums.index, autopct='%1.1f%%', startangle=140)
plt.title('Religious Composition of India')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
majority = data['Majority'].value_counts()
majority.plot(kind='bar', figsize=(10, 6))
plt.title('Majority Religion by State/UT')
plt.xlabel('Religion')
plt.ylabel('Number of States/UTs')
plt.show()
```

# Comparative Analysis

```python
data.plot(x='States/Uts', y=['Male(literacy rate)', 'Female (literacy rate)Average (literacy rate)'], kind='bar', figsize=(15, 5))
plt.title('Gender Disparity in Literacy Rates')
plt.xlabel('States/UTs')
plt.ylabel('Disparity (Male - Female)')
plt.xticks(rotation=90)
plt.legend(['Male literacy rate','Femail literacy rate'])
plt.show()
```

```python
data['Difference'] = 1000 - data['sex ratio (number of female per male)']
colors = ['blue' if diff >= 0 else 'red' for diff in data['Difference']]
data.plot(x='States/Uts', y= 'Difference', kind='bar', figsize=(15, 8), color=colors)
plt.title('Sex Ratio Analysis')
plt.xlabel('States/UTs')
plt.ylabel('Sex Ratio (females per 1000 males)')
plt.xticks(rotation=90)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)


from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color='blue', lw=4, label='More than 1000'),
                   Line2D([0], [0], color='red', lw=4, label='Less than 1000')]
plt.legend(handles=legend_elements)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Area (sq. km)', y='population(2024)', data=data, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Area vs. Population (2024)')
plt.xlabel('Area (sq. km)')
plt.ylabel('Population (2024)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data[['population(2024)', 'average (literacy rate)', 'GDP', 'per capita in', 'sex ratio (number of female per male)']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.show()
```
